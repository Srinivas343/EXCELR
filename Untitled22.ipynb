```python
import nltk
import string
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize  # Correct import
from nltk.stem import WordNetLemmatizer  # Fixed spelling
```

```python
# Download necessary NLTK resources (only run once)
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
def preprocess_text(text):
    tokens = word_tokenize(text)
    tokenize=[word.lower() for word in tokens]
    stop_words = set(stopwords.words('english'))
    tokenize=[word for word in tokenize if word not in stop_words]
    lemmatizer = WordNetLemmatizer()
    tokenize = [lemmatizer.lemmatize(word) for word in tokenize]
    return ' '.join(tokenize)
```

```python

review1 = "This is an amazing product! I love it."
review2 = "I didn't like the product. It was terrible."
```

```python

a = preprocess_text(review1)
b = preprocess_text(review2)

print(a)
print(b)
```

```python
vectorizer=TfidfVectorizer()
vectors=vectorizer.fit_transform([a,b])
print(vectors.toarray())
```

```python
cosin_sim=cosine_similarity(vectors[0],vectors[1])[0][0]
print(round(cosin_sim,3))
```

```python
def jaccab_similarity(text1, text2):
    words1 = set(text1.split())
    words2 = set(text2.split())
    intersection = words1.intersection(words2)
    union = words1.union(words2)
    if len(union) == 0:
        return 0
    jaccab_similarity = len(intersection) / len(union)
    return jaccab_similarity
```

```python
df["jaccab_similarity"] = df.apply(lambda row: jaccab_similarity(row["text1_clean"], row["text2_clean"]), axis=1)
df.head()
```
